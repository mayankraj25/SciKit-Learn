```python
# Install dependencies as needed:
# pip install kagglehub[pandas-datasets]
import kagglehub
from kagglehub import KaggleDatasetAdapter

# Set the path to the file you'd like to load
file_path = "diabetes.csv"

# Load the latest version
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "uciml/pima-indians-diabetes-database",
  file_path,
  # Provide any additional arguments like
  # sql_query or pandas_kwargs. See the
  # documenation for more information:
  # https://github.com/Kaggle/kagglehub/blob/main/README.md#kaggledatasetadapterpandas
)

df.head(15)
```

```python
# Count the number of rows where the 'Insulin' column is 0
count = (df['Insulin'] == 0).sum()
print(count)
print(df.isnull().sum())
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=0,strategy='mean')
df['Insulin']=imputer.fit_transform(df[['Insulin']])
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

X=df.drop(columns='Outcome',axis=1)
y=df['Outcome']
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.2,random_state=42)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

# Try different k values
k_range = range(1, 21)
scores = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    cv_scores = cross_val_score(knn, X, y, cv=5)  # 5-fold CV
    scores.append(cv_scores.mean())

# Best k
best_k = k_range[np.argmax(scores)]# np.argmax(scores) returns the index of the maximum value in the scores list.
print("Best k:", best_k)
```

🔍 cross_val_score()

Performs cross-validation to evaluate how well the knn model performs on the data X and labels y.

⸻

📦 What each part means:
* knn: The KNN model (e.g., KNeighborsClassifier(n_neighbors=k))
* X: Your input features (data).
* y: Your target labels (classes).
* cv=5: 5-fold cross-validation → data is split into 5 equal parts (folds).

⸻

🔄 What Happens Internally:
1.	Split the data into 5 parts (folds).
2.	For each of the 5 iterations:
	* Use 4 parts to train the model.
	* Use the remaining 1 part to test it.
3.	Repeat this so that each part is used once as test data.
4.	Return 5 accuracy scores (one from each fold) → this is cv_scores.

```python
from warnings import filterwarnings
filterwarnings('ignore')
from sklearn.metrics import accuracy_score
model=KNeighborsClassifier(n_neighbors=13)
model.fit(X_test,y_test)
predictions=model.predict(X_test)
accuracy=accuracy_score(y_test,predictions)
print(accuracy)
```
