```python
from sklearn import datasets
iris=datasets.load_iris()
X=iris.data
y=iris.target
print(X,y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
print(X_train.shape,y_train.shape)
```

```python
import pandas as pd
data=pd.read_csv('car copy.data')
data.head()
```

```python
#Extracting the required features and labes
X=data[['buying','maint','safety']].values
y=data[['class']]
print(X,y)
```

```python
import numpy as np
#The features has string valus and the ML model can work with string values , therefore converting the data
from sklearn.preprocessing import LabelEncoder

#X

#1st method
le=LabelEncoder()
for i in range(len(X[0])):
    X[:,i]=le.fit_transform(X[:,i])
print(X)

#2nd method (but put remove the .value from X)
'''['buying']=le.fit_transform(X['buying'])
X['maint']=le.fit_transform(X['maint'])
X['safety']=le.fit_transform(X['safety'])
print(X)'''

#Y

label_mapping={
    'unacc':0,
    'acc':1,
    'good':2,
    'vgood':3
}
y['class']=y['class'].map(label_mapping)
y=y['class'].values
print(y)
```

```python
#create model
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

knn=KNeighborsClassifier(n_neighbors=25,weights='uniform')
knn.fit(X_train,y_train)
prediction=knn.predict(X_test)
accuracy=metrics.accuracy_score(y_test,prediction)
#or
'''accuracy=knn.score(X_test,y_test)'''
print("Prediction: ",prediction)
print("Accuracy: ",accuracy)
```
